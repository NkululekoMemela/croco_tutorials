```python
# load modules

import numpy as np
from pydap.client import open_url
import pylab as plt
import cartopy.crs as ccrs
import cartopy
#
# thredds URL
#

DIR="http://dap.saeon.ac.za/thredds/dodsC/SAEON.EGAGASINI/2019.Penven/MONTHLY_MEAN/1_36_degree/"

file='roms_avg_Y'
suffix='_mmean.nc.2'

Y=1993
M=1

tindex=0


tout=3000

############## END OF USERS DEFINED VARS ######################################

inifile=DIR+file+str(Y)+'M'+str(M)+suffix

dataset=open_url(inifile,timeout=tout)
```

```python
## identify surface layer (= size of z-dimension/s_rho)
s_rho=dataset['s_rho']
S=np.size(s_rho)

## extract relevant lat and lon coords
## (note: lon_rho/lat_rho for temp, salt and zeta; lon_u/lat_u for u and sustr; lon_v/lat_v for v/svstr)
var=dataset['lon_rho']
lon_roms=var['lon_rho'].data[:]

var=dataset['lat_rho']
lat_roms=var['lat_rho'].data[:]

## extract mask (land = 0, ocean = 1 )
## note: mask_rho is for rho variables (see above) mask_u and mask_v for u and v variables respectively 
var=dataset['mask_rho']
mask_roms=var['mask_rho'].data[:]


## extract one timestep (given above) of sea surface temperature
var=dataset['temp']
#temp=np.squeeze(var['temp'].data[tindex,S-1,::]) ## squeeze to get rid of singleton dimensions
temp=np.squeeze(var['temp'].data[tindex,S-1,:,:])
```

```python
#### the two functions below are for converting u- and v- variables from u- and v- points to 
#### rho points. This needs to be done when calculating current speed for example, or when plotting
#### current vectors

def v2rho_2d(var_v):
    [M,Lp]=var_v.shape
    var_rho=np.zeros((M+1,Lp))
    var_rho[1:M-1,:]=0.5*np.squeeze([var_v[0:M-2,:]+var_v[1:M-1,:]])
    var_rho[0,:]=var_rho[1,:]
    var_rho[M,:]=var_rho[M-1,:]
    return var_rho

def u2rho_2d(var_u):
    [Mp,L]=var_u.shape
    var_rho=np.zeros((Mp,L+1))
    var_rho[:,1:L-1]=0.5*np.squeeze([var_u[:,0:L-2]+var_u[:,1:L-1]])
    var_rho[:,0]=var_rho[:,1]
    var_rho[:,L]=var_rho[:,L-1]
    return var_rho
```
