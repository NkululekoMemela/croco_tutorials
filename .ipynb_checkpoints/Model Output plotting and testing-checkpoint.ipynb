# ZETA (SSH Animation) Simulation

```python
time_Plot=DS['time']
s_rho_Plot =DS['s_rho']
SST_Plot =DS['temp']
```

```python
eta_rho_Plot =DS['eta_rho'].values
xi_rho_Plot =DS['xi_rho'].values
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots(figsize=(5, 3))
cax = ax.pcolormesh(xi_rho_Plot, eta_rho_Plot, zeta_Plot[0,:,:],
                    vmin=-0.05, vmax=0.06, cmap='bwr')
fig.colorbar(cax)

text = ax.text(0.5, 0.95, '', transform=ax.transAxes, ha='center')

def animate(i):
    cax.set_array(zeta_Plot[i,:,:].values.flatten())
    text.set_text(f'Time Step: {i*10}'+" days")  # Update the text header with the current time step

anim = FuncAnimation(fig, animate, frames=len(time_Plot)-1, interval=400)

plt.draw()
plt.show()

anim.save('Gambu.gif', writer='imagemagick')
```
